```python
import pandas as pd
import numpy as np

def read(path):
    datasets=pd.read_csv(path,sep=";")
    return datasets

#def pour voir les valeur manquante "NAN"

def val_manq (data):
    val_manquante=data.isna()
    return val_manquante.sum()

# suppresion des colonnes
def delet (data,col):
    return data.drop(columns=col)
```

```python
data_2013=read("trafic-annuel-entrant-par-station-du-reseau-ferre-2013.csv")
print(data_2013.columns)
```

```python
print(data_2013.shape)

val_manq(data_2013)
```

```python
col_sup=["Correspondance_2","Correspondance_3","Correspondance_4","Correspondance_5","Correspondance_1","Arrondissement pour Paris"]
delet(data_2013,col_sup)
```
