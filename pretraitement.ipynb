```python
import pandas as pd
import numpy as np
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def read(path,sep):
    datasets=pd.read_csv(path,sep=sep)
    return datasets

#def pour voir les valeur manquante "NAN"

def val_manq (data):
    val_manquante=data.isna()
    return val_manquante.sum()

# suppresion des colonnes
def delet (data,col):
    return data.drop(columns=col)
def read(path,sep):
    datasets=pd.read_csv(path,sep=sep)
    return datasets

#def pour voir les valeur manquante "NAN"

def val_manq (data):
    val_manquante=data.isna()
    return val_manquante.sum()

# suppresion des colonnes
def delet (data,col):
    return data.drop(columns=col)
```

```python
data_2013=read("trafic-annuel-entrant-par-station-du-reseau-ferre-2013.csv",sep=";")
data_2014=read("trafic-annuel-entrant-par-station-du-reseau-ferre-2014.csv",sep=";")
data_2015=read("trafic-annuel-entrant-par-station-du-reseau-ferre-2015.csv",sep=";")
data_2016=read("trafic-annuel-entrant-par-station-du-reseau-ferre-2016.csv",sep=";")
data_2017=read("trafic-annuel-entrant-par-station-du-reseau-ferre-2017.csv",sep=";")
data_2018=read("trafic-annuel-entrant-par-station-du-reseau-ferre-2018.csv",sep=";")
data_2019=read("trafic-annuel-entrant-par-station-du-reseau-ferre-2019.csv",sep=";")
data_2020=read("trafic-annuel-entrant-par-station-du-reseau-ferre-2020.csv",sep=";")
data_2021=read("Trafic annuel entrant par station du réseau ferré 2021.csv",sep=";")
```

```python
#voir les valeurs manquante dans chaque colonne du datasets
val_manq(data_2013)
```

```python
#supression des colonnes
col_sup=["Correspondance_2","Correspondance_3","Correspondance_4","Correspondance_5","Correspondance_1","Arrondissement pour Paris"]
data_2013=delet(data_2013,col_sup)
data_2014=delet(data_2014,col_sup)
data_2015=delet(data_2015,col_sup)
data_2017=delet(data_2017,col_sup)
data_2018=delet(data_2018,col_sup)
data_2019=delet(data_2019,col_sup)
data_2020=delet(data_2020,col_sup)
data_2021=delet(data_2021,col_sup)
col_sup16=["Correspondance_2","Correspondance_3","Correspondance_4","Correspondance_5","Correspondance_1","Arrondissement pour Paris","Column 12","Column 13","Column 14","Column 15"]
data_2016=delet(data_2016,col_sup16)
```

```python
#   ajouter une colone année pour tout les datasets

data_2013["Année"]=2013
data_2014["Année"]=2014
data_2015["Année"]=2015
data_2016["Année"]=2016
data_2017["Année"]=2017
data_2018["Année"]=2018
data_2019["Année"]=2019
data_2020["Année"]=2020
data_2021["Année"]=2021
```

```python
#assomnler tous les datasets
data_final=np.concatenate((data_2013,data_2014,data_2015,data_2016,data_2017,data_2018,data_2019,data_2020,data_2021),axis=0)
#data_final.to_csv('data_final.csv', index=False)
#print("Fichier CSV 'data_final.csv' créé avec succès !")
print(data_final.shape)
```

```python
# covertir le data_final en dataframe
data_final_df = pd.DataFrame(data_final, columns=data_2013.columns)
data_final_df.to_csv("data_final.csv", index=False)
print("Fichier CSV 'data_final.csv' créé avec succès !")
```

1-Évolution du trafic au fil des années (par station ou en général)

```python
# Évolution du trafic pour chaque station
plt.figure(figsize=(12, 6))
sns.lineplot(x='Année', y='Trafic', hue='Station', data=dataf ,palette='Set2')
plt.title("Évolution du trafic au fil des années (par station)")
plt.xlabel("Année")
plt.ylabel("Trafic")
plt.xticks(rotation=45)
plt.legend(title='Station', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

3-Comparaison du trafic entre les villes

```python
# Créer un graphique à barres
plt.figure(figsize=(14, 10))

# j'ai groupy le trafic pa rapport a la ville comme ça chhaque ville aura son total de trafic 

trafic_ville = dataf.groupby('Ville')['Trafic'].sum().sort_values(ascending=False)
sns.barplot(x=trafic_ville.index, y=trafic_ville.values, palette='Set2')

# Ajouter un titre et des labels
plt.title("Comparaison du trafic entre les villes", fontsize=16)
plt.xlabel("Ville", fontsize=14)
plt.ylabel("Trafic", fontsize=14)

# Rotation des labels de l'axe X pour mieux les lire
plt.xticks(rotation=45, ha="right")

# Ajuster l'espacement des éléments du graphique
plt.tight_layout()

# Affichage du graphique
plt.show()
```

4-Trafic par type de réseau (RER, Métro)

```python
#j'ai gropy le trafic par rport au type du resaux que j'ai RER metro ...
trafic_reseau = dataf.groupby('Réseau')['Trafic'].sum().reset_index()

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=trafic_reseau, x='Réseau', y='Trafic', palette='coolwarm')
plt.title("Trafic par type de réseau")
plt.xlabel('Type de réseau')
plt.ylabel('Trafic total')
plt.show()
```

5-Variation du trafic par station spécifique : évolution d'une station

```python
# specifier une station spécifique
station_specifique = dataf[dataf['Station'] == 'OPERA']

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=station_specifique, x='Année', y='Trafic', marker='o')
plt.title("Évolution du trafic de la station OPERA")
plt.xlabel('Année')
plt.ylabel('Trafic')
plt.xticks(rotation=45)
plt.show()
```
